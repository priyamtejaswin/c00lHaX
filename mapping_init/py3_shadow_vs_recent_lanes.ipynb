## Compare ...

```python
old_rows = df_2019[df_2019.num_preds>=100].shape[0]
```

```python
merged = pd.merge(df_2019[df_2019.num_preds>=100], df_2020[df_2020.num_preds>=100], 
                how='inner', suffixes=['_2019', '_2020'],
         on=[u'src_fc', u'dst_dh', u'lpe_tier', u'dg_flag'])
```

```python
print("Overlap 2019 vs 2020: %d/%d"%(merged.shape[0], old_rows)), round(merged.shape[0]*1.0/old_rows, 2)
```

```python
merged['diff'] = merged['accuracy_2019'] - merged['accuracy_2020']
```

```python
pdrop_rows = len(merged[merged['diff'] > 0])
print("Perf drop 2019 vs 2020: %d/%d"%(pdrop_rows, len(merged))), round(pdrop_rows*1.0/len(merged), 2)
```

```python
fifty_rows = len(merged[merged['diff'] >= 0.3])
print(">=30%% drop 2019 vs 2020: %d/%d"%(fifty_rows, len(merged))), round(fifty_rows*1.0/len(merged), 2)
```

```python
merged[merged['diff']>=0.3].sort_values('diff', ascending=False).head(10).to_csv('topdiff_lanesplit_accuracy.csv', index=False)
```
