Problem 2

Okay, so there is indeed a better solution to the counting problem explained on Euler.

The idea is that any number can have atmost one prime above it's square root.
- Start from 2.
- Eliminate all factors of 2.
- Take what's left and compute it's square root.
    - If the next candidate (i.e. 3) is greater than the current square root, then that's the answer.
    - Else factorize with current candidate and repeat the process.

````python
number = 103

## check for 2s
while number%2 == 0:
    number = number / 2
    
if number == 1:
    print "Answer is 2"
        
else:
    f = 3
    limit = math.sqrt(number)
    while number > 1 and f <= limit:
        if number % f == 0:
            number = number / f
            answer = f
            print "ANSWER:", answer
            ## exhaustively divide the number.
            while number % f == 0:
                number = number / f
                
            limit = math.sqrt(number)
        
            ## move to next f
        f += 2
            
    if number == 1:
        print answer
    else:
        print number, limit, f
````

## Problem 5

2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

````python
def find_remaining_factors(n, f):
    remainder = 1
    ## eliminate 2
    while f%2 == 0:
        f = f / 2
        if n%2 == 0:
            n = n / 2
        else:
            remainder *= 2

    if f == 1:
        return remainder

    else:
        p = 3
        while f > 1:
            while f%p == 0:
                f = f / p
                if n%p == 0:
                    n = n / p
                else:
                    remainder *= p
            p += 2

        return remainder
````

Okay. The very first hint is the form of primes. If I start with [2, 3], then every prime after that can be of form `6k +- 1`. So, I can jump in steps of 6 and check for +1 and -1.

````python
def stupid_check_prime(n, primes):
    isPrime = True

    for p in primes:
        if n%p == 0:
            isPrime = False
            break
            
    return isPrime
````

````python
%%timeit
primes = [2, 3]
ix = len(primes)
counter = 6
while ix < 10001:
    isPrime = stupid_check_prime(counter - 1, primes)
    if isPrime:
        ix += 1
        primes.append(counter - 1)
        
    
    isPrime = stupid_check_prime(counter + 1, primes)
    if isPrime:
        ix += 1
        primes.append(counter + 1)
        
    counter += 6
    
print primes[-1]
````

Unfortunately, the gain is minimal... probably because for every jump of 6, I am check for 2 values, whereas previously, I was checking 1 value in 3 jumps... So this is not helpful at all. 

But there is that square-root procedure which I can use : (Sieve Sieve) they keep calling it.

````python
%%timeit
ix = 1
primes = [2]
counter = 3
while ix < 10001:
    ## 2s
    isPrime = True
    limit = math.sqrt(counter)
    p_ix = 0
    while (primes[p_ix]<=limit) and p_ix<(ix):
        if counter%primes[p_ix] == 0:
            isPrime = False
            break
        p_ix += 1
            
    if isPrime:
        ix += 1
        primes.append(counter)
        
    counter += 2
    
print primes[-1]
````

Wow. This decreased significantly. From 8.7s to 425ms . Just for thoroughness, I will re-write the max-prime problem using this.

## Problem 10

The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

Okay. This seems easy enough. I'll just have to recursively apply the priming condition on every number, while also maintaining a memoized lookup of other factors.

````python
import math
````

````python
%%timeit
primes = [2]
for n in xrange(3, 2000001):
    res = check_prime(n, primes)
    if res:
        primes.append(n)
        
print sum(primes)
````

````python
def check_prime(number, prime_list):
    limit = math.sqrt(number)
    if number % 2 == 0:
        return False
    else:
        pix = 1
        while (pix < len(prime_list)) and (prime_list[pix] <= limit):
            factor = prime_list[pix]
            if number % factor == 0:
                return False
            else:
                pix += 1
                
        return True
````

````python
def erether_prime(n):
    fc = 0
    factors = []
    if n%2 == 0:
        return False
        factors.append(2)
        fc += 1
        while n%2 == 0:
            n = n/2
            
    else:
        start = 3
        limit = math.sqrt(n)
        
        while start <= limit:
            if n%start == 0:
                return False
                factors.append(start)
                fc += 1
                while n%start == 0:
                    n = n/start
                    
                limit = math.sqrt(n)
                start += 2
            else:
                start += 2
                
    if fc == 0:
        return True
    else:
        return False
````

````python
%%timeit
_sprime = 0
_lprime = 2
for n in xrange(3, 2000001):
    if erether_prime(n):
        _lprime = n
        _sprime += _lprime
        
print _lprime, _sprime
````
